```python
import csv

f=open('D:/JDY_folder/jupyter_ws/databank/card1.csv',encoding='utf-8')  # 파일 열기
data=csv.reader(f)  # 파일 읽기

next(data)  #표의 헤더 제거(제목 행)

data=list(data)
print(data)
```

```python
sum=0
for r in data:
    sum+=int(r[6])
print(f"총액: {sum:,}")
```

```python
sum=0
for r in data:
    if r[-1] == '전표매입':
        sum+=int(r[6])
        
print(f"총액: {sum:,}")
```

```python
s = {}

for row in data:
    if row[-1] == '전표매입':
        store, payment = row[5], int(row[6])
        if store not in s.keys():
            s[store] = payment
        else:
            s[store] += payment
        
print(s)
```
