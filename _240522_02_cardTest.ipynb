```python
import operator
import matplotlib.pyplot as plt

s = {}

for row in data:
    if row[-1] == '전표매입':
        store, payment = row[5], int(row[6])
        if store not in s.keys():
            s[store] = payment
        else:
            s[store] += payment
# print(s)
            
top10 = sorted(s.items(), key=operator.itemgetter(1), reverse=True)[:10]
# print(top10)

top10_store, top10_amount = [], []
for t in top10:
    top10_store.append(t[0])
    top10_amount.append(t[1])

print(top10_store)

plt.rc('font', family="Malgun gothic")
plt.title("10~12월 지출 TOP 10")
plt.barh(top10_store, top10_amount, color='b')

# current_values = plt.gca().get_xticks()
# plt.gca().set_xticklabels(['{:.0f}'.format(y) for y in current_values])

plt.show() 
```
